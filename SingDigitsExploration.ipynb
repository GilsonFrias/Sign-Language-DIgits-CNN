## Data exploration on the Sing language digits dataset

> Dataset retrieved from: https://www.kaggle.com/ardamavi/sign-language-digits-dataset

```python
import numpy as np
from skimage.util import img_as_ubyte
import skimage.io as io
import matplotlib.pyplot as plt
import os

directory = os.getcwd()+'/Sign-language-digits-dataset/'
x = np.load(directory+'X.npy')
y = np.load(directory+'Y.npy')

#The total amount of digits encoded
digits = y.shape[1] 

#Get total number of images for each digit
counts = y.sum(axis=0)
print("Images count per digit: ", counts)

#Plot a sample of each digit image
columns = int(digits/2)
rows = 2
fig, ax = plt.subplots(rows, columns, figsize=(10, 4), sharex=True)
for m in range(rows):
    for n in range(columns):
        index = m*int(x.shape[0]/2)+n*int(counts.max())
        ax[m, n].imshow(x[index])
        ax[m, n].set_axis_off()
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()

#Resize images (64, 64) -> (32, 32) and flatten them
'''
x_resized = []
for image in x:
    x_resized.append(resize(image, (32, 32)).flatten())

x = np.array(x_resized)
'''
#Figures are not stacked in order, a bit of investigation
#led to discovering the slices of the 'x' matrix that correspond
#to each digit. The 'slices' dictionary contains the interval of samples corresponding to each type of digit image.
slices = {0:np.s_[204:409], 1:np.s_[822:1028], \
          2:np.s_[1649:1855], 3:np.s_[1443:1649], \
          4:np.s_[1236:1443], 5:np.s_[1855:], \
          6:np.s_[615:822], 7:np.s_[409:615], \
          8:np.s_[1028:1236], 9:np.s_[:204]}

#Reconstruct the input set 'x' by clustering images with the same annotation together
X = np.concatenate([x[slice_] for slice_ in \
                    list(slices.values())], axis=0)


#Additionally, create a label matrix for the newly arranged data set
y = np.zeros((X.shape[0], digits), dtype='uint8')
k = np.zeros(digits, dtype='uint8')
k[0] = 1
start = 0

for digit in slices.keys():
    k = np.roll(k, -1)
    size = x[slices[digit]].shape[0]
    y[start:size+start]+=k
    start+=size
```

```python
#Save a sample of each hand sign
for n in slices.keys():
    img = x[slices[n]][0]
    io.imsave('figures/sample%d.png'%n, img_as_ubyte(img))
```
